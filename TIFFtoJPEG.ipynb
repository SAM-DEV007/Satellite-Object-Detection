```python
from osgeo import gdal
import numpy as np
```

```python
import matplotlib.pyplot as mplot
data_set1 = gdal.Open(r'/content/L3-NG43C10-094-052-09Apr14-BAND2.tif')
data_set2= gdal.Open(r'/content/L3-NG43C10-094-052-09Apr14-BAND3.tif')
data_set3= gdal.Open(r'/content/L3-NG43C10-094-052-09Apr14-BAND4.tif')
# As, there are 3 bands, we will store in 3 different variables
band_1 = data_set1.GetRasterBand(1) # red channel
band_2 = data_set2.GetRasterBand(1) # green channel
band_3 = data_set3.GetRasterBand(1) # blue channel
b1 = band_1.ReadAsArray()
b2 = band_2.ReadAsArray()
b3 = band_3.ReadAsArray()
img_1 = np.dstack((b2, b3, b1)).astype("uint8")
mplot.imshow(img_1)
mplot.show()
```
